```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

# ✅ Sample labeled dataset (you can expand it)
data = {
    "text": [
        "You're such an idiot",
        "Hi, how are you today?",
        "Let's bomb the school",
        "Have a blessed day, friend!",
        "You're a wonderful person!",
        "I'm going to hurt you",
        "I love this place"
    ],
    "label": [1, 0, 1, 0, 0, 1, 0]  # 1 = Unsafe 😡, 0 = Safe 😊
}

df = pd.DataFrame(data)

# ✅ Build pipeline
pipeline = Pipeline([
    ('vectorizer', TfidfVectorizer(ngram_range=(1, 3), max_features=500)),
    ('clf', LinearSVC(C=1.0))
])

# ✅ Train the model
pipeline.fit(df["text"], df["label"])

# ✅ Test messages
sample_texts = [
    "You're such an idiot",
    "Hi, how are you today?",
    "Let's bomb the school",
    "Have a blessed day, friend!",
    "You're a wonderful person!",
    "I'm going to hurt you",
    "I love this place"
]

predictions = pipeline.predict(sample_texts)

# ✅ Display predictions
for msg, pred in zip(sample_texts, predictions):
    label = "Unsafe 😡" if pred == 1 else "Safe 😊"
    print(f"Message: {msg}\nPrediction: {label}\n")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification


# ✅ Custom PyTorch Dataset
class ToxicDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_len=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        item = self.tokenizer(
            self.texts[idx],
            max_length=self.max_len,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )
        return {
            'input_ids': item['input_ids'].squeeze(0),
            'attention_mask': item['attention_mask'].squeeze(0),
            'labels': torch.tensor(self.labels[idx], dtype=torch.long)
        }

# ✅ Reuse your training data
texts = [
    "You're such an idiot", "Let's bomb the school", "I hate you and hope you fail",
    "I'm going to hurt you", "I can't wait to destroy everything", "I'm going to find you and kill you",
    "You're such a loser", "You're a wonderful person!", "Have a blessed day, friend!",
    "Hi, how are you today?", "You're amazing 🤗", "You're beautiful inside and out",
    "Good job on the project!", "I love this place", "Let's go for a walk"
]
labels = [1]*7 + [0]*8  # 1=Unsafe 😡, 0=Safe 😊

# ✅ Tokenizer and dataset
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")
dataset = ToxicDataset(texts, labels, tokenizer)

# ✅ DataLoader
loader = DataLoader(dataset, batch_size=4, shuffle=True)

# ✅ Model
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# ✅ Optimizer
optimizer = AdamW(model.parameters(), lr=2e-5)

# ✅ Training loop
model.train()
epochs = 5

for epoch in range(epochs):
    print(f"\n🔁 Epoch {epoch+1}/{epochs}")
    total_loss = 0
    for batch in loader:
        batch = {k: v.to(device) for k, v in batch.items()}
        optimizer.zero_grad()
        outputs = model(**batch)
        loss = outputs.loss
        total_loss += loss.item()
        loss.backward()
        optimizer.step()
    avg_loss = total_loss / len(loader)
    print(f"✅ Average training loss: {avg_loss:.4f}")
```

```python
sample_texts = [
    "You're such an idiot",
    "Hi, how are you today?",
    "Let's bomb the school",
    "You're amazing 🤗",
    "You're beautiful inside and out"
]

model.eval()
inputs = tokenizer(sample_texts, padding=True, truncation=True, return_tensors="pt").to(device)
with torch.no_grad():
    outputs = model(**inputs)
    predictions = torch.argmax(outputs.logits, dim=1)

for text, pred in zip(sample_texts, predictions):
    print(f"Message: {text}")
    print(f"Prediction: {'Unsafe 😡' if pred.item() == 1 else 'Safe 😊'}\n")
```

```python
import torch
from transformers import DistilBertTokenizerFast

# ✅ Load tokenizer & move model to correct device
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

# ✅ Test messages: 10 safe + 10 unsafe
test_messages = [
    # 🟩 Safe
    "You did a great job!",
    "I hope you have a wonderful day.",
    "Thank you so much for your help.",
    "I appreciate your effort!",
    "You're amazing 🤗",
    "Let's grab lunch together",
    "Have a blessed weekend 🙏",
    "This community is lovely!",
    "You're beautiful inside and out",
    "What a fantastic idea!",

    # 🟥 Unsafe
    "You're a piece of trash",
    "I will find you and hurt you",
    "Let's kill everyone there",
    "You are a total failure",
    "You're such a loser",
    "Just shut up already",
    "I'll bomb your house",
    "Nobody likes you",
    "You're dead meat",
    "You're the worst person I know"
]

# ✅ Tokenize
inputs = tokenizer(test_messages, truncation=True, padding=True, return_tensors="pt").to(device)

# ✅ Predict
with torch.no_grad():
    outputs = model(**inputs)
predictions = torch.argmax(outputs.logits, dim=1).cpu().numpy()

# ✅ Display results
for msg, pred in zip(test_messages, predictions):
    label = "Unsafe 😡" if pred == 1 else "Safe 😊"
    print(f"Message: {msg}\nPrediction: {label}\n")
```

```python
from sklearn.metrics import classification_report

# True labels: first 10 safe (0), next 10 unsafe (1)
true_labels = [0]*10 + [1]*10

# Print detailed metrics
print("\n📊 Classification Report:\n")
print(classification_report(true_labels, predictions, target_names=["Safe", "Unsafe"], digits=4))
```

```python
import gradio as gr
import joblib

model = joblib.load("bubble_guard_svm.pkl")

def classify_message(message):
    try:
        prediction = model.predict([message])[0]
        return "🛡️ Unsafe 😡" if prediction == 1 else "✅ Safe 😊"
    except Exception as e:
        return f"❌ Error: {str(e)}"

gr.Interface(
    fn=classify_message,
    inputs=gr.Textbox(lines=2, placeholder="Enter your message here..."),
    outputs="text",
    title="BubbleGuard demo",
    description="This model classifies text messages as Safe or Unsafe using a trained SVM model (still testing)"
).launch(share=True, debug=True)
```
